```python
import numpy as np 
import matplotlib.pyplot as plt 


x_train=np.array([1.0,2.0])
y_train=np.array([300.0,500.0])

print(f"The x_train is {x_train}")
print(f"The y_train is {y_train}")
```

```python
# m is the number of training examples 
print(f"The x_train.shape is : {x_train.shape}")
m = x_train.shape[0]
print(f"The number of training examples is :  {m}")
```

```python
for i in range(0,2):
    x_i = x_train[i]
    y_i = y_train[i]
    
    print(f"(x^({i}),(y^({i}))  = ({x_i} ,{y_i})")
```

```python
plt.scatter(x_train,y_train , marker = 'x' , c='r')
plt.title("Housing Prices")
plt.ylabel('Price (in 1000s of Dollars)')
plt.xlabel('Size (in 1000 sqft)')
plt.show()
```

```python
#  fuction is : x^(i) = wx^(i) + b

w = 100 
b = 100 
print(f"w:{w}")
print(f"b:{b}")
```

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples 
      w,b (scalar)    : model parameters  
    Returns
      y (ndarray (m,)): target values
    """
    m = x.shape[0]
    
#     zeroes ka array bana diya utna bada jitne inputs they 
    f_wb = np.zeros(m)
#     f_wb ke har index pe value update kardi 
    for i in range(m):
        f_wb[i] = w * x[i] + b
        
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b,)

# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```
